```python
# RequisiÃ§Ã£o
response = requests.get(url, headers=headers, params=querystring)
resp = response.json()

# ValidaÃ§Ã£o simples
if response.status_code == 200 and "data" in resp:
    data = resp['data']

    # Dados principais
    print(f"Nome do perfil: {data.get('name')}")
    print(f"Username: @{data.get('screenName')}")
    print(f"Link do perfil: {data.get('url')}")
    print(f"Bio: {data.get('description')}\n")

    # MÃ©tricas
    print(f"Seguidores: {data.get('usersCount')}")
    print(f"Engajamento mÃ©dio (ER): {round(data.get('avgER') * 100, 2)}%")
    print(f"Likes mÃ©dios: {data.get('avgLikes')}")
    print(f"ComentÃ¡rios mÃ©dios: {data.get('avgComments')}")
    print(f"Score de qualidade: {round(data.get('qualityScore') * 100, 1)} / 100\n")

    # AudiÃªncia
    genders = data.get('genders', [])
    ages = data.get('ages', [])
    print("DistribuiÃ§Ã£o por gÃªnero:")
    for g in genders:
        print(f"  {g['name'].capitalize()}: {round(g['percent'] * 100, 1)}%")

    print("\nDistribuiÃ§Ã£o por idade:")
    for a in ages:
        print(f"  {a['name'].replace('_', '-')} anos: {round(a['percent'] * 100, 1)}%")

    # Ãšltimos posts
    print("\nÃšltimos posts:")
    for post in data.get('lastPosts', []):
        print(f"- {post['date'][:10]} | {post['likes']} likes | {post['comments']} comentÃ¡rios")
        print(f"  Link: {post['url']}")
        print(f"  Texto: {post['text'][:100]}...\n")
else:
    print("Erro ao buscar dados ou formato inesperado.")
```

### Get account media content

```python
# Get account posts
url = "https://rocketapi-for-developers.p.rapidapi.com/instagram/user/get_media"

payload = {
	"id": 8643219911,
	"count": 12,
	"max_id": None
}
headers = {
	"x-rapidapi-key": rAPIKey,
	"x-rapidapi-host": "rocketapi-for-developers.p.rapidapi.com",
	"Content-Type": "application/json"
}

response = requests.post(url, json=payload, headers=headers)
resp = json.loads(response.text)

print(resp)
```

```python
# Visualize the structure (there is a list on items - one element for content)
resp
```

```python
# Get content items
items = resp["response"]["body"]["items"]

# Loop and store the main data
posts_data = []
for item in items:
    post = {
        "id": item.get("id"),
        "timestamp": datetime.utcfromtimestamp(item["taken_at"]).strftime('%Y-%m-%d %H:%M:%S'),
        "username": item["user"]["username"],
        "caption": item["caption"]["text"] if item.get("caption") else None,
        "image_url": item["image_versions2"]["candidates"][0]["url"] if item.get("image_versions2") else None,
        "like_count": item.get("like_count", 0),
        "comment_count": item.get("comment_count", 0),
        "permalink": f"https://www.instagram.com/p/{item['code']}/" if "code" in item else None
    }
    posts_data.append(post)

# Print out the extracted info
for post in posts_data:
    print(post)
```
